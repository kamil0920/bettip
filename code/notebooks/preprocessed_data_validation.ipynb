```python
import pandas as pd
import os

folder_path = '../datascripts/processed_data/premier_league/2020'
```

```python
if not os.path.exists(folder_path):
    print(f"Folder nie istnieje: {folder_path}")
else:
    print(f"Folder znaleziony: {folder_path}")

parquet_files = [f for f in os.listdir(folder_path) if f.endswith('.parquet')]

print(f"\nZnaleziono {len(parquet_files)} plików parquet:")
for file in parquet_files:
    print(f"  - {file}")

for file in parquet_files:
    file_path = os.path.join(folder_path, file)
    
    print(f"\n{'='*60}")
    print(f"Plik: {file}")
    print(f"{'='*60}")
    
    df = pd.read_parquet(file_path)
    
    print(f"\nKształt danych (wiersze, kolumny): {df.shape}")
    print(f"\nPierwsze 5 wierszy:")
    print(df.head())
    
    print(f"\nNazwy kolumn:")
    print(df.columns.tolist())
```
