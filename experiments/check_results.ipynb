```python
import mlflow
import pandas as pd

client = mlflow.MlflowClient()
exp = client.get_experiment_by_name("tuning-home_win-random_forest")
runs = client.search_runs(exp.experiment_id, order_by=["metrics.test_roc_auc DESC"], max_results=10)

# Pokaż top 5 wyników
top_runs = pd.DataFrame(runs)
print("Top 5 prób:")
print(top_runs[['run_id', 'data.metrics.test_roc_auc', 'data.params.max_depth', 'data.params.n_estimators']].head())
```
